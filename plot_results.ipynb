```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
with open("train_logs/tcn/2021-08-19T234405/tcn_0.0003_100classes_log.txt", "r") as f:
    file = f.readlines()
print(file)
```

```python
validation = []
for line in file:
    if 'val' in line:
        validation.append(line)
    
print(len(validation))
```

```python
import re
sample = validation[0]
RE = re.findall('\d[.\d]*', sample)
print(RE)
```

```python
val_loss = []
val_acc = []

for val in validation:
    numbers = re.findall('\d[.\d]*', val)
    val_loss.append(eval(numbers[1]))
    val_acc.append(eval(numbers[2]))
    
print('val Loss', 'val Acc')
list(zip(val_loss, val_acc))
```

```python
train = []
for line in file:
    if '100%' in line:
        train.append(line)
    
print(len(train))
```

```python
train_loss = []
train_acc = []

for t in train:
    numbers = re.findall('\d[.\d]*', t)
    train_loss.append(eval(numbers[3]))
    train_acc.append(eval(numbers[4]))
    
print('train Loss', 'train Acc')
list(zip(train_loss, train_acc))
```

```python
plt.figure(figsize=(15, 10))
plt.plot(np.array(range(73)), np.array(val_loss), label='Validation')
plt.plot(np.array(range(73)), np.array(train_loss), label='Train')
plt.legend()
plt.grid()
plt.title('Loss')
plt.savefig('LossPlot.jpg')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(np.array(range(73)), np.array(val_acc), label = "Validation")
plt.plot(np.array(range(73)), np.array(train_acc), label = "Train")
plt.legend()
plt.grid()
plt.title('Train')
plt.savefig('TrainPlot.jpg')
plt.show()
```
